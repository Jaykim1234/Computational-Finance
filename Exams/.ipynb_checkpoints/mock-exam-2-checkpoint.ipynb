## Question 1

1. **[5]** Consider the function `f(A)` below, that takes the array `A` as input. Write a docstring for the function `f(A)`. In other words, what does the function do?

```python
import numpy as np
def f(A):
    '''
    
    '''
    
    m, n = A.shape
    N = m*n
    a = np.zeros([N])
    
    for j in range(n):
        a[m*j:m*(j+1)] = A[:,j]
        
    return a
```

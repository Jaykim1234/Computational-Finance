## Question 1

1. **[5]** Consider the function `f(A)` below, that takes the array `A` as input. Write a docstring for the function `f(A)`. In other words, what does the function do?

````python
import numpy as np
def f(A):
    '''
    
    '''
    
    m, n = A.shape
    N = m*n
    a = np.zeros([N])
    
    for j in range(n):
        a[m*j:m*(j+1)] = A[:,j]
        
    return a
````

## Question 3

Let the stock price $\{S_{t}\}_{t\geq
0}$ be generated by a binomial tree with $N$ steps, where
$S_{i+1}=S_{i}u$ or $S_{i+1}=S_{i}d$, $d<e^{r\delta t}<u$. The market also contains a European-style derivative with payoff $P_T\equiv \max(S_0- S_T, 0)$, and a risk-free bond satisfying $B_t=e^{rt}$.
The risk-neutral probability is $p\equiv\mathbb{Q}[u]=(e^{r\delta t}-d)/(u-d)$ as usual, where $\delta t\equiv T/N$.

1. **[5]** Show that $$
\mathbb{E}^\mathbb{Q}\left[ S_{i+1}|\mathcal{F}_i\right]=S_{i}e^{r\delta t}.
$$
2. **[5]** What is the required minimum number of down moves of the stock in order for the derivative to expire in the money?

#### Answers

1. $\mathbb{E}^{\mathbb{Q}}[S_{i+1}\mid \mathcal{F}_i]=pS_iu+(1-p)S_id=p(u-d)S_i+dS_i=(e^{r\delta t}-d)S_i+dS_i=e^{r\delta t}S_i.$
* The derivative is equivalent to a put struck at $S_0$, which finishes in the money when the corresponding call doesn't, or when the number of up moves is smaller than or equal to  $N/2+\log(K/S_0)/(2\log u)=N/2$. Hence, we need the number of down moves to be larger than $N/2$.
    This can also be seen intuitively: the derivative finishes in the money whenever $S_T<S_0$, which happens whenever there are more down- than up moves.
 
    The above is true if $u=1/d$, which was strictly speaking not assumed in the question. While I do accept this as an answer, the question can also be answered without this assumption: let $k$ denote the number of up moves as usual, so that $N-k$ is the number of down moves. Then the option finishes in the money whenever
 \begin{align*}
  S_T=S_0u^{k}d^{N-k}&<S_0\Leftrightarrow\\
  u^{k}d^{N-k}&<1\Leftrightarrow\\
  k\log u +(N-k) \log d&<0\Leftrightarrow\\
  k(\log u-\log d) &<-N\log d\Leftrightarrow\\
  k&<-N\frac{\log d}{\log u-\log d }\Leftrightarrow\\
  N-k&>N\left(1+\frac{\log d}{\log u-\log d }\right)\Leftrightarrow\\
  N-k&>N\frac{\log u}{\log u-\log d },
  \end{align*}
which reduces to $N-k>N/2$ if $u=1/d$.

## Question 4

A researcher wants to obtain a Monte Carlo estimate of the price of a European-style option with payoff $C_T(S_T)$, written on an underlying $S_t$ that follows geometric Brownian motion with drift $r$ (the risk-free rate) under $\mathbb{Q}$. That is, its risk-neutral dynamics are

\begin{equation*}
dS_t=r S_t dt+\sigma S_t dW_t. \tag{1}\label{eq:1}
\end{equation*}

1. **[5]** Let $\delta t\equiv T/N$ for some fixed $N$. Write down a discretized version (Euler approximation) of the process in $\eqref{eq:1}$. Explain briefly how you would use the equation to simulate paths for $S_t$.
2. **[10]** Use Ito's lemma to obtain the SDE satisfied by $X_t= f(t, S_t)$, where $f(t, x)\equiv \log(x)$. Then, explain how you could use it in an alternative approach for simulating paths for $S_t$, and what the advantages in terms of accuracy and computational speed (in Python) of this approach are compared to the one in Question 4.1.
3. **[5]** Suppose the researcher has simulated $n=100$ values of $Y_t = g(S_t)$, so $Y_t$ is a function of $S_t$, and that the average over the replications for $Y_t$ is 0.56, with a standard deviation of 1.17. Compute the asymptotic confidence interval.

#### Answers

1. Writing $S_i$ for $S_{i\delta t}$,
 $$
 S_{i}=S_{i-1}+r S_{i-1} \delta t + \sigma S_{i-1}  \sqrt{\delta t} Z_i.
 $$
 Paths for $S_t$ can be simulated by drawing standard normal random numbers for $Z_i$ and iterating the above equation.
2. $X_t\equiv f(S_t)=\log S_t$, so $\dot f (S_t) = 0$, $ f' (S_t) = 1/S_t$, $f'' (S_t) = -1/S_t^2$. Note that for the term $\frac{1}{2} f'' (S_t) dS_{t}^{2}$ in Ito's Lemma we need want to determine $dS_{t}^{2}$ first. The latter follows $d S_t^{2} = r^{2} S_{t}^{2} dt^{2} + 2 r S_{t}^{2} \sigma dt dW_{t} + \sigma^{2} S_{t}^{2} dW_{t}^{2}$ where the terms $dt^{2}$ and $dt dW_{t}$ tend to zero faster than $dW_{t}^{2}$, which has variance $dt$. Setting the $dt^{2}$ and $dt dW_{t}$ terms to zero and substituting $dt$ for $dW_{t}^{2}$
 \begin{align*}
dX_t=d f(S_t)&= \dot f (S_t) dt +  f' (S_t)  dS_t + \frac{1}{2} f'' (S_t) S_t^{2} \sigma^{2} dt\\
        &= \frac{1}{S_t}  dS_t - \frac{1}{2S_t^2} S_t^{2} \sigma^{2} dt\\
        &= \frac{1}{S_t}  ( r S_t dt + \sigma S_t dW_t) - \frac{1}{2} \sigma^2 dt\\
        &=  (r-\frac{1}{2}\sigma^2) dt + \sigma dW_t .
 \end{align*}
 The alternative approach is to simulate $X_t$ rather than $S_t$, and then take the exponential. This avoids the approximation error of the Euler approximation, and it is also computationally more efficient (at least in Python), because it avoids a loop.
 
3. The confidence interval is $[\hat{g}_t \pm 1.96 \hat{\sigma}/\sqrt{n}] = [0.331,0.789]$ for $n=100$.

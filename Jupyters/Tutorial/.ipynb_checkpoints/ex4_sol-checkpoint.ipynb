## Question 1

In this exercise you will download returns on 49 industry portfolios and create sorted portfolios from them to analyze if there is an anomaly or mispricing. **1.1.** Import the relevant modules:

```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import statsmodels.api as sm
%matplotlib inline
import matplotlib.pyplot as plt
```

**1.2.** Download monthly value weighted returns on the 49 industry portfolios from [Kenneth French's website](http://mba.tuck.dartmouth.edu/pages/faculty/ken.french/data_library.html) for the period January 2010 to September 2020.

Download the data set `'49_industry_portfolios'` from source `'famafrench'` via `pandas-datareader`. The output of DataReader will be a `dict` object ([documentation](https://docs.python.org/3/library/stdtypes.html?highlight=dict#dict)). Use `dict['DESCR']` to see the data description. Index into the `dict` object to select the monthly value weighted returns and store it in a new DataFrame: use `dict[key]`, where `key` will be the index number as given in the data description.

*Note:* The column headers are all strings of length 5. An additional space is added if the industry name has less than 5 letters, e.g. 'Food ' instead of 'Food'. The spaces can be removed using the following code: `R.columns = [x.strip(' ') for x in R.columns]`, assuming that your DataFrame is called `R`. The function `strip(' ')` removes any leading and trailing spaces. It is not necessary to do this for the exercise, but avoids unexpected behavior when trying to select a column.

```python
startdate = pd.datetime(2010,1,1)
enddate = pd.datetime(2021,9,30)
#enddate = pd.datetime(2020,12,30)
ds = web.DataReader('49_industry_portfolios', 'famafrench', start=startdate, end=enddate)
print(ds['DESCR'])
```

```python
R = ds[0]  # Average value weighted returns; monthly
R.columns = [x.strip(' ') for x in R.columns]  # Remove spaces in column names
R.tail()
```

**1.3.** We will need size information to compute value weighted portfolio returns. We will calculate size per industry as the product of the average firm size in the industry times the number of firms in the industry. Both, average size and number of firms dataframes, are available in the same `dict` object as the returns. Also store size in a new DataFrame. 

```python
size = ds[5] * ds[4]  # Average size times number of firms
size.columns = [x.strip(' ') for x in size.columns]  # Remove spaces in column names
size.tail()
```

**1.4.** Before sorting and grouping the assets into portfolios, we need to compute or read the characteristic(s). Momentum is defined as the return from month $t-12$ to $t-2$.

```python
r = np.log(1 + R / 100) * 100  # Transform to log returns, such that we can aggregate over time
mom = r.shift(2).rolling(11).sum()  # Computes the sum over the period t-12 to t-2.
mom = mom[12:]
mom.head()
```

**1.5.** Now that we have a characteristic, we can form the sorted portfolios. Create $P=7$ portfolios sorted on your momentum. Compute the portfolio returns and store them in a new DataFrame `R_portf`.

*Note*: If you have portfolios sorted on momentum or reversal, these are resorted and grouped each month. If you have portfolios based on accouting numbers, these are resorted annually.

Below is the implementation in the form of a function.

```python
def portsort(R, X, P, size, w='vw'):
    '''Computes P portfolio returns sorted on one characteristic X, with monthly rebalancing.'''
    
    # Preliminaries
    R_portf = pd.DataFrame(columns=range(1,P+1), index=R.index, dtype='float')   # Empty DataFrame, but columns of group numbers and index of r, and float dtype
    q = np.linspace(0,1,P+1)  # quantile breakpoints on [0,1] interval
    
    # For each month
    for t in R.index:
        X_t = X.loc[t]  # Define for convenience
        
        # Find bounds/breakpoints as quantiles and convert to array (remove index)
        brkpts = X_t.quantile(q).values  
        brkpts[0] = -np.Inf   # First lower bound
        
        for p in range(1,P+1):  # Note: df[p] will index to portfolio p in DataFrame with integer column headers, but list[p] to the element indexed by p in a list/array.
            # Select assets and time period
            R_smpl = R.loc[t, (brkpts[p-1] < X_t) & (X_t <= brkpts[p])]
            
            if w == 'vw':
                # Compute the value weighted return
                size_smpl = size.loc[t, (brkpts[p-1] < X_t) & (X_t <= brkpts[p])]  # Select size
                R_portf.loc[t, p] = R_smpl@size_smpl / size_smpl.sum()  # Sum-product with relative size
            elif w == 'ew':
                R_portf.loc[t, p] = R_smpl.mean()  # Equal weighted return
            else:
                raise ValueError('Only equally weighted ("ew") and value weigthed ("vw") supported.')

    return R_portf#.astype('float')  # Ensure that contents are floats
```

```python
P = 7
R_portf = portsort(R[12:], mom, P, size = size[12:], w = 'vw')
R_portf.head()
```

**1.6.** Compute the average return over time for each portfolio. Is there a pattern?

```python
R_portf.mean()
```

Portfolios with higher momentum have on average higher returns over the sample period. This is particularly strong when comparing portfolio 1 with portfolio 6 and 7. Momentum is a characteristic that distinguishes between assets in terms of returns. Portfolio 2 is a bit of an outlier though.

**1.7.** Now that we have our portfolios, we can set up the factor model. The monthly factors of the Fama-French three factor model (FF3) can be obtained from [Kenneth French's website](http://mba.tuck.dartmouth.edu/pages/faculty/ken.french/data_library.html). 

Download the data set `'F-F_Research_Data_Factors'` from source `'famafrench'` via `pandas-datareader`. The output of DataReader is again a `dict` object.

```python
ds2 = web.DataReader('F-F_Research_Data_Factors','famafrench',start=startdate,end=enddate)
print(ds2['DESCR'])
```

```python
ff3factors = ds2[0]
ff3factors = ff3factors[12:]
ff3factors.tail()
```

**1.8.** Obtain the portfolios' excess returns by deducting the risk-free rate ('RF' in the DataFrame of the FF3 factors) from the portfolios returns.

```python
R_portf = R_portf.subtract(ff3factors.RF, axis=0)
R_portf.dropna(inplace=True)
R_portf.tail()
```

**1.9.** Estimate the $\alpha_p$ and $\beta_p$ by using the following $P$ time series regressions:

$$
R_{pt}-R_{f,t} = \alpha_p + \beta_p' f_t + \varepsilon_{pt},
$$

where $\varepsilon_{pt} \sim N(0,\sigma^2_p)$, for $t = 1,\ldots,T$, $p = 1,\ldots,P$, where $f_t$ are your risk factors. In the case of the FF3 model, we have $f_t = (R_{mt}-R_{f,t},SMB_{t},HML_{t})'$.

Store the alpha estimates and $t$ statistics in a new DataFrame.

```python
ff3factors = sm.add_constant(ff3factors.drop(columns='RF'))  # Add a constant, and remove risk free rate
ff3factors.head()
```

```python
alpha = []
t_a = []
ff3ret = []  # Expected return
for p in range(1,P+1):
    ff3 = sm.OLS(R_portf[p], ff3factors)
    results = ff3.fit()
    alpha.append(results.params[0])
    t_a.append(results.tvalues[0])
    ff3ret.append(results.fittedvalues.mean() - results.params[0])
```

```python
R_abnorm = pd.DataFrame(index=range(1,P+1))
R_abnorm['alpha'] = alpha
R_abnorm['t'] = t_a 
```

**1.10.** Inspect the alphas and $t$ statistics. Are there significant abnormal returns?

```python
R_abnorm
```

There is little statistical power to reject the null of $\alpha_p = 0$ for any of the portfolios. At least in this sample it seems that the FF3 model explains the differences in average returns across portfolios.

**1.11.** Make a scatter plot of the average return against the expected return (as implied by the asset pricing model).

```python
plt.figure(figsize=(8,6))
plt.plot(ff3ret, R_portf.mean(), marker='o')
for i, p in enumerate(R_portf.columns):   # enumerate(x) returns iterable with each element a tuple: (index, x[index])
    xi = ff3ret[i]
    yi = R_portf[p].mean()
    plt.annotate(p, (xi, yi), xytext=(xi, yi+0.01))   # Label the portfolios
plt.xlabel('$E[R_{pt}-R_{f,t}]$')
plt.ylabel('$R_{pt}-R_{f,t}$')
plt.title('Realized Versus Expected Return (FF3)');
```

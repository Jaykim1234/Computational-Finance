```python
mu = df_ret.mean()
Sigma = df_ret.cov()
std = df_ret.std()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(std, mu);
plt.xlim(0,5)
plt.xlabel(r'$\sigma$')
plt.ylabel(r'$\mu$')
plt.title('Mean and volatility of Euro STOXX 50 constituents');
```

```python
def MV_RiskyPortfolio(mu, Sigma, mu_p):
    '''Computes the weights and variance of the mean-variance efficient portfolio with expected return mu_p.'''
    
    e = np.ones(mu.shape)  # N x 1
    Sigma_inv = np.linalg.inv(Sigma)
    A = mu.T @ Sigma_inv @e
    B = mu.T @ Sigma_inv @mu
    C = e.T @ Sigma_inv @ e
    D = B*C - A**2
    
    wp = (mu_p*C - A)/D * Sigma_inv@mu + (B - A*mu_p)/D*Sigma_inv@e
    sigma2_p = wp.T @ Sigma @ wp
    
    return wp, sigma2_p
```

```python
mu_target = 0.1
wp, sigma2_p = MV_RiskyPortfolio(mu, Sigma, mu_target)
print(f'Optimal variance of mean-variance portfolio for a given expected returns of {mu_target}% is: {sigma2_p.round(2)}')
print(f'Optimal weights:\n{pd.Series(wp, index=mu.index)}')
```

```python
mu_y = np.linspace(-0.8, 0.8, 100)
sigma_x = [np.sqrt(MV_RiskyPortfolio(mu, Sigma, m)[1]) for m in mu_y]  
# If a function has multiple output variables, you can immediately index into that: 
# f(x)[i] will select the output of f(x) corresponding to index i.
```

```python
# Figure
plt.figure(figsize=(8,6))
plt.plot(sigma_x, mu_y)
plt.xlim(0,5)
plt.scatter(std, mu, color='k')
plt.xlabel(r'$\sigma$')
plt.ylabel(r'$\mu$')
plt.legend(['MV Frontier','Stocks'])
plt.title('Mean-Variance Efficient Frontier');
```
